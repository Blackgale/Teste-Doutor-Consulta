## Leitura rápida
A ideia aqui é sair de uma tabela crua e chegar a uma leitura útil para compras, financeiro e gestão.


```python
metrics.groupby('commodity')['price_usd_ton'].describe().round(2)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
for commodity, group in metrics.groupby('commodity'):
    plt.plot(group['date'], group['price_usd_ton'], label=commodity)
plt.legend()
plt.title('Séries históricas')
plt.tight_layout()
plt.show()
```
